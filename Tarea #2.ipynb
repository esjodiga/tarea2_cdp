# Version control: What is Git?
## ¿Qué es Git?

-  Git es un software de control de versiones. Las herramientas de control de versiones permiten la trazabilidad de los cambios en el código de algún proyecto. Adicional al almacenamiento de los cambios en los contenidos también guardan metadata de ellos, pudiendo verificar con ello quien realizó un cambio, a qué archivos y cuándo lo hizo.
- Estas herramientas son muy útiles tanto cuando se trabaja en lo individual como en un trabajo colaborativo dentro de un proyecto de desarrollo de software.
-  Git no solo permite trabajar con versiones de un proyecto en una línea temporal recta, sino que también hace posible trabajar con diferentes versiones de forma paralela (ramas o _branches_ en inglés) para luego unir (_merge_ en inglés) las diferentes variantes del proyecto en una sola con todas las características.


## ¿Cómo funciona Git?

Cuando es indicado por parte del usuario Git almacena un _snapshot_, imagen o versión del proyecto que se trabaja. Luego, crea una serie de referencias y nombres internos que diferencian y organizan las versiones entre sí.

De forma predeterminada Git almacena las versiones en una branch principal (llamada _master_), una referencia lineal. Dependiendo de los comandos que sean indicados por el usuario, Git puede crear nuevas branches de forma paralela a la principal y luego puede unir versiones de diferentes branches en una sola versión de ser necesario. Algunos comandos básicos para usar Git son:
1. git **clone**: Para clonar un repositorio remoto en nuestra computadora (a nivel local).
2. git **branch**: Para trabajar con las ramas
3. git **checkout**
4. git **status**: Nos informa información sobre la rama en que se encuentra el usuario.
5. git **add**: Para incluir loscambios realizados en el paquete que Git guardará.
6. git **commit**: Uno de los más usados, para guardar cambios.
7. git **push**: Para _enviar_ los cambios realizados al repositorio remoto.
8. git **pull**: Para _recibir_ actualizaciones de un repositorio remoto.
9. git **revert**: Para deshacer cambios.
10. git **merge**: Para unir dos ramas.

Información sobre comandos extraída de: __[Freecodecamp](https://www.freecodecamp.org/espanol/news/10-comandos-de-git-que-todo-desarrollador-deberia-saber/)__

## ¿Y GitHub?
GitHub es una plataforma para alojar proyectos de código como complemento a la herramienta de Git. A través de diferentes comandos dentro de Git, GitHub (copia remota) interactúa con nuestra computadora (copia local).

## Conclusiones y comentarios:
-  Git es una importante herramienta para el desarrollo de código fuente en la actualidad, dado que no solo permite la trazabilidad de los cambios que se van realizando en el proyecto sino permite el almacenamiento de versiones o _snapchats_ por si en caso aparece algún problema, sea más fácil regresar a una de estas versiones anteriores.
-  Esta herramienta de control de versiones también hace más fácil el trabajo colaborativo, dado que varias personas pueden trabajar en un mismo proyecto al mismo tiempo y ramificar las versiones de forma individual para luego unirlas nuevamente más adelante.
-  Github es una herramienta potente y amigable para combinar con Git, a pesar de que no es la única, es de las más utilizadas.
